```python
#小数点数を演算するときの気になる現象
num = 0
num += 0.1
print(num) #0.1
num += 0.1
print(num) #0.2
num += 0.1
print(num) #0.30000000000000004
num += 0.1
print(num) #0.4
num += 0.1
print(num) #0.5
num += 0.1
print(num) #0.6
num += 0.1
print(num) #0.7
num += 0.1
print(num) #0.7999999999999999
num += 0.1
print(num) #0.8999999999999999
num += 0.1
print(num) #0.9999999999999999
```
