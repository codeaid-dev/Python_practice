```python
#super()の実行
class Car:
    def __init__(self, type):
        self.type = type
    def run(self):
        print("ブッブー")

class Ferrari(Car):
    def __init__(self, type, color):
        super().__init__(type)
        self.color = color
    def run(self):
        print("ビューン")

car = Ferrari("スポーツカー", "赤")
print(f'車種: {car.type} 色: {car.color}') # 車種: スポーツカー 色: 赤
```

```python
#Exercise6
class Calc:
    def add(self, a, b):
        return a+b
    def multiply(self, a, b):
        return a*b

class Work:
    def __init__(self):
        self.work = Calc()
    def combine(self,a,b):
        '''aとbをCalcクラスのadd()とmultiply()の結果を足して返す'''
        return self.work.add(a,b) + self.work.multiply(a,b)

n = Work()
print(n.combine(10,5))
```

```python
#Exercise7
class Time:
    def __init__(self, minutes=0, seconds=0):
        self.minutes = minutes
        self.seconds = seconds
    def __str__(self):
        return f'{self.minutes:02d}分{self.seconds:02d}秒'
    def __add__(self, add_sec):
        new_time = Time()
        new_time.minutes = self.minutes
        new_time.seconds = self.seconds
        new_time.seconds += add_sec

        if new_time.seconds >= 60:
            new_time.minutes += new_time.seconds // 60
            new_time.seconds = new_time.seconds % 60
        
        return new_time

print(Time(4,5) + 150)
```

```python
#Exercise8
import random
class Dice():
    def __init__(self, sides=6):
        self.sides = sides

    def roll_dice(self):
        return random.randint(1, self.sides)

d = Dice()
results = []
for num in range(10):
    result = d.roll_dice()
    results.append(result)
print(f"{d.sides}面サイコロを10回転がした結果")
print(results)
rate = results.count(3) / len(results) * 100
print(f'3のでる確率は{rate}%でした')
```

```python
#Exercise9-1
class Dessert:
    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients
    def has_ingredient(self, ingredient):
        return ingredient in self.ingredients
    @staticmethod
    def get_kinds():
        return ('ドリンク', 'アイス')

coffee = Dessert('コーヒー', ['コーヒー豆', '水'])
latte = Dessert('カフェラテ', ['コーヒー豆', '水', '牛乳'])
icecream = Dessert('アイスクリーム', ['牛乳', '卵', '生クリーム', '砂糖'])
for ing in ('茶葉', 'コーヒー豆', '牛乳', '生クリーム'):
    if coffee.has_ingredient(ing):
        print(f'{coffee.name}は{ing}を含みます')
    if latte.has_ingredient(ing):
        print(f'{latte.name}は{ing}を含みます')
    if icecream.has_ingredient(ing):
        print(f'{icecream.name}は{ing}を含みます')
print(f'Dessertクラスには、{Dessert.get_kinds()}の種類があります')
```

```python
#Exercise9-2
class ComboDessert(Dessert):
    def __init__(self, name, entrees):
        super().__init__(name, entrees)
    def has_ingredient(self, ingredient):
        for entree in self.ingredients:
            if entree.has_ingredient(ingredient):
                return True
        return False

coffee = Dessert('コーヒー', ['コーヒー豆', '水'])
icecream = Dessert('アイスクリーム', ['牛乳', '卵', '生クリーム', '砂糖'])
afoguard = ComboDessert('アフォガード', [coffee, icecream])
print('アフォガードは', end='')
for s in afoguard.ingredients:
    print(s.ingredients, end='')
print('を含みます')
if (afoguard.has_ingredient('牛乳')):
    print('アフォガードは牛乳を含みます')
```

```python
#Exercise12-1
import random

class Character:
    def __init__(self, name, hp):
        self.name = name
        self.hp = hp
    def attack(self, character):
        damage = random.randint(20,25)
        character.hp = character.hp - damage
        return damage
```

```python
#Exercise12-2
class Hero(Character):
    def __init__(self, name, hp):
        super().__init__(name, hp)
        self.__skill = 40
    def sp_attack(self, monster):
        monster.hp = monster.hp - self.__skill
        return self.__skill
```

```python
#Exercise12-3
class Monster(Character):
    def __init__(self, name, hp):
        super().__init__(name, hp)
    def heal(self):
        heal = random.randint(25,50)
        self.hp = self.hp + heal
        return heal
```

```python
#Exercise12-4
hero = Hero('勇者', 100)
monster = Monster('スライム', 120)
turn = True
while hero.hp > 0 and monster.hp > 0:
    if turn: #Hero turn
        if random.randint(1,4) == 1:
            damage = hero.sp_attack(monster)
            print(f'{hero.name}のスペシャル攻撃！！')
        else:
            damage = hero.attack(monster)
        print(f'{hero.name}は{monster.name}に{damage}のダメージを与えた')
        turn = False
    else: #Monster turn
        if monster.hp <= 10:
            heal = monster.heal()
            print(f'{monster.name}は{heal}回復した')
        else:
            damage = monster.attack(hero)
            print(f'{monster.name}は{hero.name}に{damage}のダメージを与えた')
        turn = True

if hero.hp > 0 and monster.hp <= 0:
    print(f'{hero.name}は{monster.name}をやっつけた')
elif hero.hp <= 0 and monster.hp > 0:
    print(f'{hero.name}は{monster.name}にやられてしまった・・・')
else:
    print(f'{monster.name}は逃げてしまった')
print(f'{hero.name}のHP:{hero.hp}\n{monster.name}のHP:{monster.hp}')
```
