```python
import tkinter
import tkinter.messagebox
count = 0
def timer():
    global count
    print(count)
    if count == 5:
        tkinter.messagebox.showinfo('タイマー','時間切れ！')
    else:
        root.after(1000, timer)
    count += 1

root = tkinter.Tk()
root.title('タイマーテスト')
timer()
root.mainloop()
```
