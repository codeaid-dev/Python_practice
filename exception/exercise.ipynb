```python
def check_odd(x):
    if x%2==0:
        raise ValueError('引数の値は偶数です')
    else:
        print('引数の値は奇数です')
        print(f'x:{x%2==1}')

try:
    check_odd(1)
except ValueError as err:
    print(err)
```

```python
class MyExceptionA(Exception):
    pass

def check_even(x):
    if x%2==1:
        raise MyExceptionA('引数の値は奇数です')
    else:
        print('引数の値は偶数です')
        print(f'x:{x%2==0}')
try:
    check_even(1)
except MyExceptionA as err:
    print(err)
```

```python
class MyExceptionA(Exception):
    pass
class MyExceptionB(MyExceptionA):
    pass

def check_even(x):
    if x%2==1:
        raise MyExceptionA('引数の値は奇数です')
    elif x>=6:
        raise MyExceptionB('値が6以上です')
    else:
        print('引数の値は5以下で偶数です')
try:
    check_even(6)
except MyExceptionA as err:
    print(err)
except MyExceptionB as err:
    print(err)
```

```python
def pickup(l, i):
    try:
        print(l[i])
    except IndexError:
        print('インデックスが範囲外です')
    else:
        print('アクセスできました。', f'長さ{len(l)}, インデックス{i}')

pickup([1,2,3], 2)
```

```python
def pickup(l, i):
    try:
        print(l[i])
    except IndexError:
        print('インデックスが範囲外です')
    else:
        print('アクセスできました。', f'長さ{len(l)}, インデックス{i}')
    finally:
        print('必ず実行します')

pickup([1,2,3], 3)
```

```python
# Exercise1
try:
    x = int(input('1つ目の整数: '))
    y = int(input('2つ目の整数: '))
    result = x / y
    print(f'{x}/{y}={result}')
except ZeroDivisionError:
    print('0で割り算できません')
except ValueError:
    print('数字を入力してください')
```
