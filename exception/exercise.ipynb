```python
def check_odd(x):
    if x%2==0:
        raise ValueError('引数の値は偶数です')
    else:
        print('引数の値は奇数です')
        print(f'x:{x%2==1}')

try:
    check_odd(1)
except ValueError as err:
    print(err)
```

```python
class MyExceptionA(Exception):
    pass

def check_even(x):
    if x%2==1:
        raise MyExceptionA('引数の値は奇数です')
    else:
        print('引数の値は偶数です')
        print(f'x:{x%2==0}')
try:
    check_even(1)
except MyExceptionA as err:
    print(err)
```

```python
class MyExceptionA(Exception):
    pass
class MyExceptionB(MyExceptionA):
    pass

def check_even(x):
    if x%2==1:
        raise MyExceptionA('引数の値は奇数です')
    elif x>=6:
        raise MyExceptionB('値が6以上です')
    else:
        print('引数の値は5以下で偶数です')
try:
    check_even(6)
except MyExceptionA as err:
    print(err)
except MyExceptionB as err:
    print(err)
```

```python
def pickup(l, i):
    try:
        print(l[i])
    except IndexError:
        print('インデックスが範囲外です')
    else:
        print('アクセスできました。', f'長さ{len(l)}, インデックス{i}')

pickup([1,2,3], 2)
```

```python
def pickup(l, i):
    try:
        print(l[i])
    except IndexError:
        print('インデックスが範囲外です')
    else:
        print('アクセスできました。', f'長さ{len(l)}, インデックス{i}')
    finally:
        print('必ず実行します')

pickup([1,2,3], 3)
```

```python
# Exercise7
class OutOfRangeError(Exception):
    pass

try:
    n = int(input('1~100の範囲で整数を入力: '))
    if not (1 <= n <= 100):
        raise OutOfRangeError
    print(f'入力された値: {n}')
except OutOfRangeError:
    print('1~100の範囲で入力してください')
except ValueError:
    print('整数を入力してください')
```

```python
# Exercise8
data = [10,20,30,40]
index = input('取り出すインデックス：')
def pickup(data, index):
    try:
        i = int(index)
        return data[i]
    except Exception:
        raise
try:
    result = pickup(data, index)
    print(f'取り出すデータ：{result}')
except Exception as err:
    print(f'例外発生：{err}')
```

```python
# Exercise10
import random

def do_task():
    num = random.randint(1,10)
    if 1 <= num <= 7:
        raise Exception

retries = 3
while retries > 0:
    try:
        do_task()
        print('タスク成功')
        break
    except Exception:
        retries -= 1
        print(f'ランダムエラーが発生：残りリトライ回数({retries})')
        if retries == 0:
            print('すべてのリトライに失敗しました')
```
