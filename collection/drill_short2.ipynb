```python
#Excercise6
dic = {"name": "Tanaka",
      "age":25, 
      "salary": 150000, 
      "city": "Tokyo"}
keys = ["name", "salary"]

new_dic = dict()
for k in keys:
   new_dic[k] = dic[k]
print(new_dic)

#内包表記
new_dic = {k: dic[k] for k in keys}
print(new_dic)
```

```python
#Excercise13
pair = ['Q', 'A']
num = 4
result = []
for y in range(1,num+1):
    for x in pair:
        result.append(f'{x}{y}')
print(result)

#内包表記
result2 = [f'{x}{y}' for y in range(1,num+1) for x in pair]
print(result2)
```

```python
#Excercise16
#width,height = map(int, input('文字数と行数：').split(','))
nums = input('文字数と行数：').split(',')
width,height = int(nums[0]),int(nums[1])
words = ['o', 'x']
order = 1
for i in range(height):
    moji = ''
    order = 1 - order
    for j in range(width):
        if j % 2 == 0:
            moji += words[order]
        else:
            moji += words[1 - order]
    print(moji)
```
