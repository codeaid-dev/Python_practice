```python
#Exercise6
w1 = input('文字列入力 >>')

w2 = ''
while len(w2) == 0 or len(w2) > len(w1):
    w2 = input('数える文字列 >>')

count = 0
for i in range(len(w1)):
    if w1[i:i+len(w2)] == w2:
        count += 1
print(f'"{w1}"にある"{w2}"の数は、{count}個')
```

```python
#Exercise9
s = ''
for i in range(1, 31):
  if i%3 == 0 and i%5 == 0:
    s += "FizzBuzz"
  elif i%3 == 0:
    s += "Fizz"
  elif i%5 == 0:
    s += "Buzz"
  else:
    s += str(i)
  s += ' '
print(s)

for i in range(1, 31):
  if i%3 == 0 and i%5 == 0:
    print("FizzBuzz", end=' ')
  elif i%3 == 0:
    print("Fizz", end=' ')
  elif i%5 == 0:
    print("Buzz", end=' ')
  else:
    print(i, end=' ')
```

```python
#Exercise10
w1,w2 = '',''
while True:
    w1 = input('入力1 >>')
    w2 = input('入力2 >>')
    if len(w1) > 0 and len(w2) > 0 and len(w1) == len(w2):
        break
    else:
        print('空入力はなく、同じ文字数を入力してください')
cnt = 0
for i in range(len(w1)):
    if w1[i] == w2[i]:
        cnt += 1
print(f'同じ文字は：{cnt}個')
```

```python
#Exercise11
num = ''
total, count = 0, 0
while not 'end' == num:
    num = input('入力 >>')
    if not num.isdigit():
        if 'end' != num:
            print('整数を入力してください')
        continue
    total += int(num)
    count += 1
print(f'{count}回入力・合計{total}・平均{total/count}')
```

```python
#Exercise13
letter = 'Beautiful is better than ugly.'
word = input('入力 >>')
for char in letter:
    if word == char:
        print(f'{word}を見つけました')
        break
else:
    print(f'{word}は見つかりませんでした')

pos = 0
while pos < len(letter):
    if word == letter[pos]:
        print(f'{word}を見つけました')
        break
    pos += 1
else:
    print(f'{word}は見つかりませんでした')


if word in letter:
    print(f'{word}を見つけました')
else:
    print(f'{word}は見つかりませんでした')
```

```python
#Exercise20
s = input('入力 >>')
result = ''
index = 0
end = len(s)
while index < end:
    if s[index] == ':':
        result += s[index-1] * (int(s[index+1])-1)
        index += 2
    else:
        result += s[index]
        index += 1
print(result)
```

```python
#Exercise21
num = int(input('入力 >>'))
total = 0
digit = 2
formula = ''
for i in range(num):
    formula += f'{digit} + '
    total += digit
    digit = digit * 10 + 2
print(f'{formula[:-3]} = {total}')
```

```python
#Exercise22
num = ''
while not num.isdigit():
    num = input('>>')
num = int(num)
num2 = num * 9
print(f'{num} * 9 = {num2}')
nums = str(num2)
result = 0
while True:
    for s in nums:
        result += int(s)
    if len(str(result)) == 1:
        break
    else:
        nums = str(result)
        result = 0
print(f'{num2}の各桁を足すと最後は{result}になる')
```
